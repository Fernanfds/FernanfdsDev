Exercicio : Gráfico de Juros Compostos com Python

Fórmula Basica dos Juros Compostos: M = P * (1+i)^t

```python
import matplotlib.pyplot as plt
import numpy as np

P = float(input("Digite o Valor Inicial (R$): "))
i = float(input("Qual a Taxa de Juros por Periodo (em %)? ")) /100
t = int(input("Qual o Numero de Periodos? "))

# Gerar os Periodos (Meses, Anos, etc.)

periodos = np.arange(0, t + 1 )

# Calcular o Montante ao longo do Tempo

M = P * (1 + i) ** periodos

# Plotar o Grafico

plt.figure(figsize= (10,6))
# Changed marker style to 'o' which represents a circle
plt.plot(periodos, M, label=f"Juros Compostos", color="green", linestyle="-", marker="o")

# Personalizar o Grafico
plt.title("Grafico de Juros Compostos")
plt.xlabel("Periodos")
plt.ylabel("Montante (R$)")
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.show()
```
