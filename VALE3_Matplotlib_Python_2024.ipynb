```python
import pandas as pd
import plotly.express as px
import numpy as np

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.options.display.float_format = '{:,.2f}'.format
```

```python
df = pd.read_csv("/content/VALE3 .csv")
df.head()
```

```python
df.tail()
```

```python
df['FECHAMENTO'] = df['FECHAMENTO'].str.replace(',', '.').astype(float)

df.groupby("ABERTURA").agg({"FECHAMENTO": np.mean}).sort_values("FECHAMENTO", ascending=False).reset_index()
```

```python
df_agrupado_FECHAMENTO = df.groupby("ABERTURA").agg({"FECHAMENTO": np.mean}).sort_values("FECHAMENTO", ascending=False).reset_index()
```

```python
df_agrupado_FECHAMENTO
```

```python
# @title FECHAMENTO

from matplotlib import pyplot as plt
df_agrupado_FECHAMENTO['FECHAMENTO'].plot(kind='hist', bins=20, title='FECHAMENTO')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
px.bar(df_agrupado_FECHAMENTO, x='ABERTURA', y='FECHAMENTO')
```

```python
df_agrupado_FECHAMENTO = df.groupby(["MÍNIMO", "MÁXIMO"]).agg({"FECHAMENTO": np.mean}).sort_values("FECHAMENTO", ascending=False).reset_index()
```

```python
df_agrupado_FECHAMENTO
```

```python
px.bar(df_agrupado_FECHAMENTO, x='MÍNIMO', y='FECHAMENTO', barmode = 'group')
```

```python
px.scatter(df, x='MÍNIMO', y='FECHAMENTO', hover_name = 'VOLUME')
```
