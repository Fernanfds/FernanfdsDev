```python
acoes = {

    'ticker' : ['BBAS3', 'VALE3','CMIG4','ROMI3','FLRY3', 'PETR4', 'RANI3', 'CXSE3'],
    'p/l' : ['4,63', '5,15', '5,76', '7,11', '14,93', '6,12', '12,66', '12,90'],
    'dy' : ['8,84%', '12,48%', '15,45%', '6,60%', '5,75%', '16,92%', '6,76%', '8,79%'],
    'cagr_5_anos' : ['17,16%', '9,12%', '10,60%', '10,54%', '19,42%', '10,54%', '11,31%', '7,21%']
}





df = pd.DataFrame(acoes)
df
```

```python
# @title ticker vs p/l

from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
plt.subplots(figsize=(8, 8))
df_2dhist = pd.DataFrame({
    x_label: grp['p/l'].value_counts()
    for x_label, grp in df.groupby('ticker')
})
sns.heatmap(df_2dhist, cmap='viridis')
plt.xlabel('ticker')
_ = plt.ylabel('p/l')
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
import plotly.express as px # Import Plotly Express for interactive plots

plt.subplots(figsize=(8, 8))
df_2dhist = pd.DataFrame({
    x_label: grp['dy'].value_counts()
    for x_label, grp in df.groupby('ticker')
})
sns.heatmap(df_2dhist, cmap='viridis')
plt.xlabel('ticker')
_ = plt.ylabel('dy')

# Convert the matplotlib figure to a Plotly figure
fig = px.imshow(df_2dhist.values,
                labels=dict(x="Ticker", y="DY", color="Count"),
                x=df_2dhist.columns,
                y=df_2dhist.index
               )
fig.show()  # Display the interactive plot
```
